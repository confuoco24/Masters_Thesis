```python
import random
import sympy

# Step 1: Find a prime number "p" and a generator "g"
def generate_large_prime(bits):
    return sympy.randprime(2**(bits-1), 2**bits)

# Choose the desired bit length (e.g., 1024 bits)
bit_length = 1024
p = generate_large_prime(bit_length)
print(f"Generated prime p: {p}")


def find_generator(p): # Finding g for p
    for g in range(2, p):
        if pow(g, p-1, p) == 1:
            return g

g = find_generator(p)
print(f"Chosen generator g: {g}")



def generate_data(p, g, M):
    data = []

    # Step 2: Choose s uniformly at random
    s = random.randint(0, p - 2)

    # Step 3: Perform operations for each data sample
    for _ in range(M):
        # Draw each sample z uniformly at random
        z = random.randint(0, p - 2)

        # Store (z, +1) or (z, -1) based on a condition
        label = 1 if s <= z <= (s + (p - 3) // 2) else -1

        # Compute x = g^z mod p for every sample in M
        x = pow(g, z, p)

        # Pair the result x with the earlier label
        data.append((x,z,s, label))

    return data

M = 300 # Number of samples desired
generated_data = generate_data(p, g, M)
print(generated_data)
```
